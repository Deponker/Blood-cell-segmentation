```python
import cv2
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import tensorflow as tf
from tensorflow import keras as K
import skimage
from skimage.io import imshow, imread, imsave
from glob import glob
import os
```

```python
in_dir = 'C:/Users/Shazid-PC/Desktop/Shazid repo/Original'
out_dir = 'C:/Users/Shazid-PC/Desktop/Shazid repo/k-mean'

# read the input image file names with paths into a list
infiles = in_dir + '/*.jpg'
img_names = glob(infiles)
print(img_names)
```

```python
for fn in img_names:
    print('processing %s...' % fn)

    img = cv2.imread(fn)
    img2 = img.reshape((-1,3))
    img2=np.float32(img2)
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    k=3
    attempts=10
    ret,label,center=cv2.kmeans(img2,k,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
    center=np.uint8(center)
    res = center[label.flatten()]
    res2 = res.reshape((img.shape))


    # write the result to disk in the previously created output directory
    name = os.path.basename(fn)
    outfile = out_dir + '/' + name
    cv2.imwrite(outfile, res2)
```
